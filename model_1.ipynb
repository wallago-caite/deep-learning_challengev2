## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df = application_df.drop(columns = ['EIN', 'NAME'])
application_df
```

```python
# Determine the number of unique values in each column.
application_df.nunique()
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
app_val_count = application_df['APPLICATION_TYPE'].value_counts()

app_val_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 300

application_types_to_replace = app_val_count[app_val_count<cutoff_value].index

# Replace in dataframe
for app in application_types_to_replace:
    application_df['APPLICATION_TYPE'] = application_df['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
class_val_count = application_df['CLASSIFICATION'].value_counts()

class_val_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
class_val_count[class_val_count>1]
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_value_class = 100

classifications_to_replace = class_val_count[class_val_count<cutoff_value_class].index

# Replace in dataframe
for cls in classifications_to_replace:
    application_df['CLASSIFICATION'] = application_df['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df['CLASSIFICATION'].value_counts()
```

```python
# Convert categorical data to numeric with `pd.get_dummies` --- creates booleans of the categories
application_df = pd.get_dummies(application_df)

application_df
```

```python
# Split our preprocessed data into our features and target arrays
# note that features are x = everything except price or whatever target, in this case what is successful?
# note that target array are y = price or target everything but the "successful column"
X = application_df.drop(columns='IS_SUCCESSFUL').values
y = application_df['IS_SUCCESSFUL'].values

# Split the preprocessed data into a training and testing dataset -- choose to stratify so that the classes are more balanced out bc it's a small dataset and imbalanced
X_train, X_test, y_train, y_test = train_test_split(X, y,stratify=y,test_size=.2, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
nn = tf.keras.models.Sequential()

input_shape = X.shape[1] # number of columns in the dataset

# First hidden layer
nn.add(tf.keras.layers.Dense(units=128, activation='relu', input_shape=(input_shape,))) # first layer inputting shape to avoid keras error message

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=64, activation='relu'))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model

# define optimizer -- these can be adam(go-to), SGD(basic), RMSProp(RNN,online learning), Adagrad(sparse data, text processing), adadelta(better adagrad)
# define loss function -- these can be binary crossentropy, categorical crossentropy, mean squared or error (continuous data), hinge loss
# define metrics -- for classification like this--- this is usually "accuracy"

# Compile the model
nn.compile(optimizer='adam',
           loss='binary_crossentropy',    # Loss function: Binary crossentropy for binary classification
           metrics=['accuracy'])
```

```python
# Train the model
nn.fit(X_train_scaled,
       y_train,
       epochs = 10,
       batch_size = 32,
       shuffle = True,
       verbose = 2,
       validation_data=(X_test_scaled, y_test))
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
